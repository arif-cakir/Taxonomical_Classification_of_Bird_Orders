### Dataset Explanation
The dataset contains of 18 columns. first 6 are taxonomical properties of the birds:
- "class": Taxonomic class of the bird (which is same for all birds).
- "order": Taxonomic order of the bird.
- "family": Taxonomic family of the bird.
- "genus": Taxonomic genus of the bird.
- "spp": Taxonomic species of the bird.
- "species": Subject's taxonomical name, which is "genus"_"spp".

After that, columns 6-13 are for estimated values of physiological and ethological features of the birds:
- "clutch.size": Species' estimated size of clutch.
- "max.longevity": Maximum age of species.
- "incubation": Species' estimated duration of incubation
- "dev.mode": Development model of species. "A" for altricial, "SA" for semialtricial, "P" for precocial and "SP" for semiprecocial.
- "fledling.age": Estimated age of fledging.
- "brain.mass": Estimated brain mass of species.
- "egg.mass": Species' estimated egg mass.
- "body.mass": Species' estimated body mass
- "specimen.origin": Origin of species; captivity and wild.

Finally, last three columns are for reference issues and not significant for the project:
- "qc": Quality control acceptability. Only acceptable ones will be taken for the project.
- "sample.size": size of the sample; small, medium, large.
- "references": Used references from website. 

## Purpose of the Project
There are two objectives in this project: Firstly, it is aimed to taxonomically classify bird species in respect of their physiological and ethological characteristics and secondly, creating new labels as close as possible to taxonomic labels with the help of clustering algorithms. In biological taxonomy, organisms are classified by their basis of shared physical likeness with the help of eight hierarchical categories which are: domain, kingdom, phylum, class, order, family, genus, and species. This hierarchy goes from general to specific and all classes are an inclusive group of higher ranks. Moreover, it is planned to verify that physiological and ethological characteristics like egg size, incubation duration, development model, etc. can be used as an indicator of taxonomic classification in this project. Aves class contains 27 orders, 98 families, and 338 genera; due to the higher number of families and genera, orders are going to be used in this project.

## Methods
Due to the fact that one of the objectives is classifying and other one is clustering, both supervised and unsupervised algorithms are going to be used. Firstly, Support Vector Machines will be used to classify birds. Secondly, due to the bottom-up approach of the second objective, agglomerative hierarchial clustering is going to be used. Furthermore, additional models will be constructed by DBSCAN and k-means algorithms for comparison with the first model. Finally, the clusters are going to be named.
